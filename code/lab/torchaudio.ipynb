```python
import torch
import torchaudio
import requests
from matplotlib import pyplot
```

```python
# url = "https://pytorch.org/tutorials/_static/img/steam-train-whistle-daniel_simon-converted-from-mp3.wav"
# r = requests.get(url)

file = "./data/steam-train-whistle-daniel_simon-converted-from-mp3.wav"

# with open('file', 'wb') as f:
#     f.write(r.content)



waveform, sample_rate = torchaudio.load(file)

print("Shape of waveform: {}".format(waveform.size()))
print("Shape of waveform: {}".format(sample_rate))

pyplot.figure()
pyplot.plot(waveform.t().numpy())

```

```python
specgram = torchaudio.transforms.Spectrogram()(waveform)
speclog = specgram.log2()
specloggram = specgram.log2()[0,:,:]

print("Shape of spectrogram: {}".format(specgram.size()))
print("Shape of speclog: {}".format(speclog.size()))
print("Shape of specloggram: {}".format(specloggram.size()))


pyplot.figure()
pyplot.imshow(specloggram.numpy(), cmap='gray')
```
