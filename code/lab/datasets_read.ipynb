```python
import torch
import matplotlib.pyplot as plt
import torchvision
import torchvision.transforms as transforms

imgnet_train = torchvision.datasets.QMNIST('~/Datasets/QMNIST', train=True, download=True, transform=transforms.ToTensor())

imgnet_test = torchvision.datasets.QMNIST('~/Datasets/QMNIST', train=False, download=True, transform=transforms.ToTensor())

imgnet_pil = torchvision.datasets.QMNIST('~/Datasets/QMNIST', train=False)



print(len(imgnet_train), len(imgnet_test))

print(imgnet_train[0][1])

images = []
labels = []

for i in range(10):
    images.append(imgnet_train[i][0])
    labels.append(imgnet_train[i][1])

_, ax = plt.subplots(1, 10, figsize=(20, 20))

for x, img, label in zip(ax, images, labels):
    x.imshow(img.view(28, 28))
    x.set_title(label)
    x.axes.get_xaxis().set_visible(True)
    x.axes.get_yaxis().set_visible(True)









```
