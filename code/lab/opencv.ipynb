```python
import cv2
import matplotlib.pyplot as plt 
import pylab

pylab.rcParams['figure.figsize'] = (12.0, 10.0)

# >0=0<0alpha
RGB_Image = cv2.imread('data/walle.png',1)
b, g, r = cv2.split(RGB_Image)
#  opencv  B G R  matplotlib 
#  merge 
RGB_Image = cv2.merge([r,g,b])
#
plt.subplot(1,4,1) # row, col, index
plt.imshow(RGB_Image,'gray')
plt.title('RGB_Image')
plt.subplot(1,4,2)
plt.imshow(r,'gray')
plt.title('R_Channel')
plt.subplot(1,4,3)
plt.imshow(g,'gray')
plt.title('G_Channel')
plt.subplot(144)
plt.imshow(b,'gray')
plt.title('B_Channel')

plt.show()
```
