```python
import cv2
import matplotlib.pyplot as plt 
import pylab

pylab.rcParams['figure.figsize'] = (12.0, 10.0)

# 第二个参数>0原图，=0灰度图，<0带alpha通道
RGB_Image = cv2.imread('data/walle.png',1)
b, g, r = cv2.split(RGB_Image)
# 注意在 opencv 中为 B G R ，利用 matplotlib 显示需要转换
# 使用 merge 函数将其通道进行融合
RGB_Image = cv2.merge([r,g,b])
#显示各通道信息
plt.subplot(1,4,1) # row, col, index
plt.imshow(RGB_Image,'gray')
plt.title('RGB_Image')
plt.subplot(1,4,2)
plt.imshow(r,'gray')
plt.title('R_Channel')
plt.subplot(1,4,3)
plt.imshow(g,'gray')
plt.title('G_Channel')
plt.subplot(144)
plt.imshow(b,'gray')
plt.title('B_Channel')

plt.show()
```
