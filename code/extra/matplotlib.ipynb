```python
from matplotlib import pyplot as plt
import numpy as np

# 直接画图
plt.plot([1, 2, 3, 4], [1, 4, 2, 3])
```

```python
x = np.linspace(0, 2, 100)
# print(x)
# 先画子图
fig, axs = plt.subplots()
axs.plot(x, x, label='line')
axs.plot(x, x**2, label='square')
axs.set_xlabel('x')
axs.set_ylabel('y')
axs.set_title('table')
axs.legend() # 增加左上角标牌

```

```python
# 画多个图
fig, (ax1, ax2) = plt.subplots(1, 2)
ax1.plot(x, x, label='line')
ax2.plot(x, x ** 3, label='dada')
```
